#### get genres for top reviewed apps with rating more than 4.5

```python
df_cleaned = my_data_cleaned.drop_duplicates(subset='App')
# Filter for apps with rating more than 4.5
filtered_df = df_cleaned [df_cleaned['Rating'] > 4.5]
filtered_df
```

```python
print(filtered_df['Reviews'].dtype)
filtered_df['Reviews'] = pd.to_numeric(filtered_df['Reviews'], errors='coerce')
print(filtered_df['Reviews'].dtype)
```

```python
top_data = filtered_df.sort_values("Reviews", ascending=False )
top_data
```

```python
# Initialize a set to track unique genres
unique_genres = set()
count = 0 
# Initialize an empty list to store the top genres
top_10_unique_genres = []

# Iterate over the rows of the sorted DataFrame
for index, row in top_data.iterrows():
    genre = row['Genres']
    count+=1
    # Add the genre to the set if it is not already there
    if genre not in unique_genres:
        unique_genres.add(genre)
        top_10_unique_genres.append(genre)
        
    
    # Stop when we've found 10 unique genres
    if len(unique_genres) == 10:
        print(count)
        break

# Print out the list of the top 10 unique genres
print(top_10_unique_genres)
```

```python
# Check how many unique genres are present
unique_genres_count = top_data['Genres'].nunique()

print(f"Number of unique genres: {unique_genres_count}")
```

```python
top_data_100=top_data.head(100)
top_data_100["Genres"].value_counts()
```

```python
# Create your displot
# sns.set_theme(style="ticks", color_codes=True,font_scale=1.3)
g = sns.displot(top_data_100, x="Genres", height=20, aspect=1.5)
plt.xticks(rotation=90)
# Set the font size for specific elements
g.set_axis_labels("Genres", "Frequency", fontsize=22)  # Axis labels
g.fig.suptitle("Distribution of Genres", fontsize=22)  # Title
g.set_xticklabels(g.ax.get_xticklabels(), fontsize=24)  # X-axis tick labels
g.set_yticklabels(g.ax.get_yticklabels(), fontsize=26)  # Y-axis tick labels

# Show the plot
plt.show()
```
